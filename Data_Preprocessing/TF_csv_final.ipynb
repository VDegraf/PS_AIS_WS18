## 1. Hilfsfunktionen zum Berechnen von TF

```python
import csv
import ast

def read_csv(path): # Funktion zum Einlesen der gefilterten csv Datei
    with open(path, "r") as csv_file:
        reader = csv.reader(csv_file, delimiter=';')
        liste = []
        for row in reader:
            liste.extend(row)
    return liste

def get_wordSet(liste): # Funktion zum Erstellen einer wordSet liste mit allen Wörter  Kategorie 
    i = 2
    wordSet = []
    while len(liste)>i:
        element = ast.literal_eval(liste[i]) # entfernt String
        wordSet.extend(element) # fügt das Element in die neue Liste ein
        i+=2
    return wordSet

def get_wordDict_abs(wordSet): # Funktion zum Erstellen der wordDict mit Absoluten Häufigkeiten als dict
    wordDict = dict.fromkeys(wordSet, 0)
    for word in wordSet:
        wordDict[word]+=1
    return wordDict

def computeTF(wordDict, wordSet): # Funktion zum Berechnen der relativen Häüfigkeiten als dict
    tfDict = {}
    wordSetCount = len(wordSet)
    for word, count in wordDict.items():
        tfDict[word] = count/float(wordSetCount)
    return tfDict
```

## 2. Funktionen zum Berechnen der relativen und absoluten Vorkommenshäufigkeit.

```python
def tfDict_abs(path): # Gesamtfunktion zum Einlesen, Berechnung des absoluten TFs und Ausgabe als dict
    liste = read_csv(path)
    wordSet = get_wordSet(liste)
    wordDict = get_wordDict_abs(wordSet)
    return wordDict

def tfDict_rel(path): # Gesamtfunktion zum Einlesen, Berechnung des relativen TFs und Ausgabe als dict
    liste = read_csv(path)
    wordSet = get_wordSet(liste)
    wordDict = get_wordDict_abs(wordSet)
    tfDict = computeTF(wordDict, wordSet)
    return tfDict
```

## 3. TF Ordner anlegen

```python
import os
if os.path.exists("../TF"):
    print("Ordner Data_filtered existiert bereits")
if not os.path.exists("../TF"):
    os.makedirs("../TF")
    print("Ordner Data_filtered wurde erfolgreich angelegt")
```

## 4. "TF....csv" erstellen und in dem TF Ordner abspeichern
#### Gesamtfunktion zum einlesen der gefilterten CSV Datei, TF Berechnung und Erstellung der "TF....csv" Datei in dem "TF" Ordner

```python
import time
import csv
def tf_rel(path, name):
    if os.path.isfile("../TF/%s.csv" %name):
        return print("Die %s.csv existiert bereits" %name)
    else:
        start_time = time.time()
        print("CSV wird angelegt")
        tfDict = tfDict_rel(path)
        with open("../TF/%s.csv" %name, 'w') as f:
            w = csv.writer(f)
            w.writerows(tfDict.items())
        end_time = time.time()
        return print("Datei erfolgreich angelegt \n" 'Dauer: {:5.3f}min'.format((end_time-start_time)/60))
```

#### Input: path = Pfad der gefilterten CSV Datei (nach preprocessing Schritten), name = Name nue zu erstellenden TF....csv Datei

```python
tf_rel("../Data_filtered/reviews_Musical_Instruments.csv", "TF_reviews_Musical_Instruments")
```
